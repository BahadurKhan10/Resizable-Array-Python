```python
class resizable_array:
    def __init__(self,arraysize):
        #size of array a/c to our input / creating an array
        self.arraysize = arraysize
        self.array = [None for i in range(self.arraysize)]  
        self.temp =1
    
    def __insert(self,value):
        for i in range(len(self.array)):
            #If array is empty, add value to it
            if self.array[i] == None:
                self.array[i] = value
                break

        #print(self.array)
        return self.array

    def insert_at_end(self, value):
        #condition 1: if self.array is empty
        if None in self.array:
                self.__insert(value)
        
        #condition 2: if self.array has values
        else:
            self.newsize= 1+len(self.array)
            self.newarray = [None for i in range(self.newsize)] #resizing +1 size
            for i in range(len(self.array)):
                self.newarray[i]=self.array[i]
                # hold has value of i
                hold=i

            self.newarray[hold+1]=value #put value in i+1
            self.array = self.newarray
            self.newarray=None
            #print(self.array)
        return self.array
        
    
    def printarray(self):
        print(self.array)
    
    def lengtharray(self):
        print(len((self.array)))
        
    def delete_all(self):
        x=self.array.clear()
        print(x)

    def insert_at_first(self, value):
        #create new array of +1 size from current array
        #simply add value at 0 (inserting in first index)
        
        self.size_newarray = len(self.array)+1
        self.newarray = [None for i in range(self.size_newarray)]
        self.newarray[0] = value
        
        for val in range(1,self.size_newarray):
            self.newarray[val] = self.array[val-1]
        self.array = self.newarray
        self.newarray = None
        
        #print(self.array)
        #return self.array

    def delete_from_end(self):
    #1st condition: if array has some values 
        try:
        #3rd condition if array is empty
            if len(self.array)==0:
                print("Array is empty")
                
            elif len(self.array)!=0:
                self.newsize = len(self.array)-1 #newarray -1 size of current array
                self.newarray = [None for i in range(self.newsize)]
            
                for i in range(len(self.array)-1):
                    self.newarray[i] = self.array[i]
            
                self.array=self.newarray
                #print(self.array)
    
    #2nd condition: if len of array is equal to 1
            elif len(self.array)==1:
                self.array =[]
        except:
            pass
   


    def delete_from_first(self):
        try:
        #1st condition: if array has one element then directly return empty array
            if len(self.array)==1:
                self.array =[]
                
        #2nd condition: if array has more than one element    
            elif len(self.array)!=0 and len(self.array)!=1:
                self.size_newarray = len(self.array) - 1
                self.newarray = ['insert' for i in range(self.size_newarray)]
                
                for i in range(0,len(self.newarray)):
                    self.newarray[i] = self.array[i+1]
                self.array = self.newarray
                self.newarray = None
                #print(self.array)
                return self.array
            
        #3rd condition: if array is empty
            else:
                print('Array is empty!')
        except:
            pass
            


    def insert_anywhere(self, insertafter, value):
        self.size_newarray = len(self.array) + 1
        self.newarray = [None for i in range(self.size_newarray)]
        
    #1st condition if last element is insertafter
        #simply call insert_at_end function
        if self.array[len(self.array)-1] == insertafter:
            self.insert_at_end(value)
        
    #2nd condition if insertafter is anywhere except last
        elif insertafter in self.array:
            
            for i in range(0, len(self.array)-1):
                #if insert after==i (in that loop)
                #than add value at i+1
                if self.array[i] == insertafter:
                    self.newarray[i] = self.array[i]
                    self.newarray[i + 1] = value
                    i = i + 2
                    break

                else:
                    self.newarray[i] = self.array[i]
        
            for j in range(i , len(self.newarray)): #here i=i+2 means it will range of last remaining values
                self.newarray[j] = self.array[j - 1]

            self.array = self.newarray
            self.newarray = None
            #print(self.array)
        else:
            print("the value after which you want to add, does not exists in Array")

    def delete_anywhere(self,value):
        self.create_new_size = len(self.array)-1
        self.newarray = [None for i in range(self.create_new_size)]
        
        #Basic condition
        if len(self.array)==0:
            print("Array is empty!")
        else:
            for i in range(len(self.array)):
                try:
                    #1st condition; if i not equal to value
                    if self.array[i]!=value:
                        self.newarray[i]=self.array[i]
                        
                    #2nd condition; if i equal to value #skip that i (for that loop) and take next value(i+1)
                    elif self.array[i]==value:
                        self.newarray[i]=self.array[i+1]
                except:
                    pass
            self.array=self.newarray
            self.newarray=None
            #print(self.array)
    
```
